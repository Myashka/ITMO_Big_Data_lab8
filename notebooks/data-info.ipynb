```python
not_na_df = df[
    [
        "code",
        "product_name",
        "energy-kcal_100g",
        "energy_100g",
        "fat_100g",
        "saturated-fat_100g",
        "trans-fat_100g",
        "cholesterol_100g",
        "carbohydrates_100g",
        "sugars_100g",
        "fiber_100g",
        "proteins_100g",
        "salt_100g",
        "sodium_100g",
        "calcium_100g",
        "iron_100g",
        "nutrition-score-fr_100g",
        "categories_en"
    ]
].dropna()
```

```python
not_na_df["energy-kcal_100g"] = not_na_df["energy-kcal_100g"].astype(float)
```

```python
not_na_df.sample(1000).to_csv('../data/openfood.csv', index=False)
```
